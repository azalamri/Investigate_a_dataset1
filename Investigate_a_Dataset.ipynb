### Research Question 1 (What year has the highest number of released  movies??)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.



data=df.groupby('release_year').count()['id']
print(data.tail())

#make group of the data according to their release year and count the total number of movies in each year and pot.
df.groupby('release_year').count()['id'].plot(xticks = np.arange(1960,2016,5))

#set the figure size and labels
sns.set(rc={'figure.figsize':(15,10)})
plt.title("Comparsion betweent Year and the Number Of Movies" ,fontsize = 24)
plt.xlabel('Release year',fontsize = 22 )
plt.ylabel('Total of number Of Movies',fontsize = 22)
sns.set_style("whitegrid")
```

### Research Question 2  (What was the highest movie profit??)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
df['Profit'] = df['revenue'] - df['budget']

info = pd.DataFrame(df['Profit'].sort_values(ascending = False))
info['original_title'] = df['original_title']
data = list(map(str,(info['original_title'])))
x = list(data[:10])
y = list(info['Profit'][:10])

#make a plot using pointplot for top 10 profitable movies.
ax = sns.pointplot(x=y,y=x)

#setup the figure size
sns.set(rc={'figure.figsize':(20,20)})
#setup the title and labels of the plot.
ax.set_title("the best top profitable movies ",fontsize = 22)
ax.set_xlabel("Profit, billions",fontsize = 22)
sns.set_style("darkgrid")
```
